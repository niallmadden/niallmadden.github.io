# CS319 - Week 6: Convergence of a Quadratre Rule

In this Jupyter notebook, we want to verify the convergence of the Trapezium Rule.

We'll use the `numpy` and `matplotlib` modules.

```python
import numpy as np
import matplotlib.pyplot as plt
```

We start by copying some data computed by `05CheckConvergence.cpp`

```python
N=np.array([4, 8, 16, 32, 64, 128, 256, 512])
T=np.array([8.940076e-03, 2.236764e-03, 5.593001e-04, 1.398319e-04, 3.495839e-05, 8.739624e-06, 2.184908e-06, 5.462270e-07])
```

```python
plt.plot(N, T, 'o')
plt.show()
```

That last figure is not very informative...
* on the horizontal axes, the data are too spread out
* on the vertical axis the data is too compressed near $T=0$.

We resolve this by using logarithmic axes:

```python
plt.loglog(N, T, 'o')
plt.show()
```

Looks like a straight line!

We'll use the (depreciated) `np.polyfit()` function to compute the coefficients of the line that best fits (in a least squares sense) the points `(log(N), log(T))`.

```python
q1 = np.polyfit( np.log(N), np.log(T),1)
print(q1)
```

This returns the coefficients of best-fit line.

Let's plot to check:

```python
Ns = np.linspace(N[0], N[-1], 100)
plt.plot(np.log(N), np.log(T), 'o', np.log(Ns), np.log(Ns)*q1[0]+q1[1], '--')
plt.show()
```

```python
plt.loglog(N, T, 'o', N, np.exp(q1[1])*N**q1[0], '--')
plt.show()
```
